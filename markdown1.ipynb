# 연습문제

아래의 코드의 빈 줄을 완성하여 날짜와 시간이 실행 결과와 같이 출력되도록 구현해보세요

```python
year = '2021'
month = '02'
day = '03'
hour = '18'
minute = '33'
second = '23'

print(year, month, day, ______________________)
print(hour, minute, second, __________________)
```

>실행결과  
>2021/02/03 18:33:23

````python
year = '2021'
month = '02'
day = '03'
hour = '18'
minute = '33'
second = '23'

print(year, month, day, end=" ",sep="/")
print(hour, minute, second, sep=":")
````
