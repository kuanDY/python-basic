# 데이터 시각화

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
plt.figure() # 그래프를 그리기 위한 틀을 만든다
plt.axes() # 그 안에 좌표평면들을 그린다.axes([왼쪽,아래,폭,높이])
plt.show() # 실제 화면에 표시한다
plt.subplot() #그리드 상에 그래프를 배치, 쓰는 붭  subplot(행의 수, 열의 수, 플롯번호)
```

```python
plt.plot(np.arange(2,7))
plt.show()
```

```python
plt. figure(figsize=[10,5])

plt.plot(np.arange(5)) # 파란선
plt.plot(np.arange(2,7)) #노란선
plt.plot(np.arange(4,8)) # 초록선
plt.plot(np.arange(6,10)) #빨간선

plt.show()
```

```python
fig , ax = plt.subplots(2,2,figsize=[10,5])
print(ax.shape)     #  왼쪽 위부터 오른쪽아래로 순서대로 1,2,3,4로표현
    
ax[0,0].plot(np.arange(5))   #1. 파란선
ax[0,0].plot(np.arange(2,7)) #1. 초록선
ax[0,0].plot(np.arange(1,5)) #1. 노란선
ax[0,1].plot(np.arange(5))   #2. 파란선
ax[0,1].plot(np.arange(2,6)) #2. 노란선
ax[1,0].plot(np.arange(5,8)) #3. 노란선
ax[1,0].plot(np.arange(3,5)) #3. 파란선
ax[1,1].plot(np.arange(5))   #4.파란선

plt.show()
```

```python
arr = np.arange(5)
arr1 = np.arange(2,6)
plt.figure(figsize=[5,5])

plt.subplot(221)
plt.plot(arr) # 1

plt.subplot(222)
plt.plot(arr) #2
plt.plot(arr1)

plt.subplot(223)
plt.plot(arr) #3

plt.subplot(224)
plt.plot(arr) #4
plt.plot(arr1)
```

```python
a = np.arange(10)
b = np.arange(5,10)
plt.figure(figsize=[5,5])

for x in a, b:
    if plt.subplot(222):
        print(plt.plot(a))
    elif plt.subplot(223):
        print(plt.plot(a))
    else:
        print(x)
```

# 곡선

```python
fig , ax = plt.subplots(1,2,figsize=(10,5))
fig.suptitle("hohohohoho")

x = range(0,10)
y = np.exp(x)
ax[0].plot(x,y)
ax[0].set_title("exp")
ax[0].set_xlabel("x axis")
ax[0].set_ylabel("y aixs",rotation=90, labelpad=10)

x = range(1,10000)
y = np.log(x)
ax[1].plot(x,y)
ax[1].set_title("log")
ax[1].set_xlabel("X axis")
ax[1].set_ylabel("Y axis",rotation=90, labelpad=10)
plt.show()
```

```python
x = range(1,1000)
y = np.log(x)

fig, ax = plt.subplots()
ax.plot(x,y)

ax.tick_params(axis="y",labelrotation=90)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=[10,5])
ax.plot(range(50))

ax.set_xticks(range(50))
ax.set_yticks(range(50))

ax.set_ylabel("Y axis",labelpad=10)

plt.show()
```

```python
y = np.exp(range(0,10))
fig , ax = plt.subplots(2,2,figsize=[10,5])

ax[0,0].plot(y,marker="o")
ax[0,1].plot(y,marker="v")
ax[1,0].plot(y,marker="x")
ax[1,1].plot(y,marker="^")

plt.show()
```

```python
x = np.arange(0,4,0.5)

plt.figure(figsize=(10,5))

plt.plot(x, x+1 , 'bo')
plt.plot(x, x**2-4, 'g--')
plt.plot(x, -2*x+3, 'r:')

plt.axvline(1.0, 0.1, 0.8, color='lightgray', linestyle='--', linewidth=10)
plt.vlines(1.8, -3.0, 2.0, color='gray', linestyle='solid', linewidth=3)

plt.show()
```

```python
x = np.arange(5)
years = ['2018','2019','2020','2021','2022']
values = [400,600,800,1000,1200]

plt.bar(x,values)
plt.xticks(x,years)

plt.show()
```

색깔 바꾸기

```python
x = np.arange(5)
years = ['2018','2019','2020','2021','2022']
values = [400,600,800,1000,1200]

plt.bar(x,values,color='gray')
plt.xticks(x,years)

plt.show()
```

막대선 사이즈(width)

```python
x = np.arange(5)
years = ['2018','2019','2020','2021','2022']
values = [400,600,800,1000,1200]

plt.bar(x,values,width=0.1)
plt.xticks(x,years)

plt.show()
```

축 바꾸기

```python
x = np.arange(5)
years = ['2018','2019','2020','2021','2022']
values = [400,600,800,1000,1200]

plt.barh(x,values,color='gray')
plt.xticks(x,years)

plt.show()
```
