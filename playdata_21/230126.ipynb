```python
# 파라미터를 따로 지정 안하면 초기값이 들어감
# 파라미터 중 오른쪽 파라미터가 기본값으로 지정되지 않았으면 왼쪽 파라미터는 기본값을 지정할 수 없다

# global - 전역 변수 접근 제어자
def get_stock(code = None, s_page = None, e_page = None):
    res = []
    for _ in range(s_page, e_page):
        url = "https://finance.naver.com/item/sise_day.naver?code={}&page={}".format(code, _)
        r = requests.get(url, headers = head)
        bs = BS(r.text).find("table", {"class" : "type2"}).find_all("span", "tah")
        day = []
        for o in bs:
            day.append(o.text.strip())
            if(len(day) == 7):
                res.append(day)
                day = []
    return res
        
get_stock("035720", 1, 3)
```

```python
from bs4 import BeautifulSoup as BS
import os
from tqdm import trange, notebook  # 진행률을 확인할 수 있는 모듈
os.makedirs("./stocks", exist_ok = "True")
# 반복문에 tqdm을 사용하여 진행률을 GUI로 표시할 수 있다.
for target in notebook.tqdm(r.json()['OutBlock_1']):
    res = []
    f = open("./stocks/{}.csv".format(target['ISU_ABBRV']), "w", encoding = "utf-8", newline = "")
#     print(target['ISU_ABBRV'])
    url = "https://finance.naver.com/item/sise_day.naver?code={}&page=1".format(target['ISU_SRT_CD'])
    r = requests.get(url, headers = head)
    bs = BS(r.text, "lxml").find("table", {"class" : "type2"}).find_all("span", "tah")
    for _ in bs:
        res.append(_.text.strip())
        if(len(res) == 7):
            break
    wr = csv.writer(f)
    wr.writerow(res)
    f.close()
```
