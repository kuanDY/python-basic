```python
import pandas as pd
```

```python
items = root.iter(tag='item')
total = []
for element in items:
    dict_ = {}
    for x in element:
        dict_[x.tag] = x.text
    total.append(dict_)
    
```

# decorator

함수안에 함수를 정의 함 - 기존 함수에서 추가하고 싶은 기능을  추가하는 함수

```python
def func_a(func):
    def inner():
        print("inner function")
        func()
    return inner
```

```python
@func_a
def func_b():
    print("func_b start")
```

```python
func_b()
```
